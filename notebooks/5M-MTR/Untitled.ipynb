```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import os
```

```python
import matplotlib.pyplot as plt
```

```python
mtr_df = pd.read_csv('ft_results_combined.csv')
```

```python
mtr_df
```

```python
molnet_mean_cols = mtr_df.filter(regex=r'test\S*mean').columns.tolist()
```

```python
sns.set_context('talk')
fig, axes = plt.subplots(len(molnet_mean_cols)//2, 2, figsize=(20, 30), sharex=True)
for i in range(len(molnet_mean_cols)):
    sns.regplot(data=mtr_df, x='min_eval_loss', y=molnet_mean_cols[i], ax=axes.flatten()[i])
fig.subplots_adjust(hspace=0.6)
plt.savefig('mtr_means.png', dpi=200)
```
