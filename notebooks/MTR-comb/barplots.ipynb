```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
import os
```

```python
csv_paths = [os.path.join('../', x, 'ft_results_all_seeds.csv') for x in ["5M-MTR", "10M-MTR", "77M-MTR"]]
```

```python
csv_paths
```

```python
dfs = [pd.read_csv(p) for p in csv_paths]
```

```python
# only pick run_names from 77
run_names_to_use = dfs[-1].run_name.unique()
```

```python
mtr_df = pd.concat(dfs, axis=0).reset_index(drop=True).sort_values(by='min_eval_loss')
mtr_df = mtr_df[mtr_df.run_name.isin(run_names_to_use)]
```

```python
mtr_df.to_csv('mtr_results_all_seeds.csv', index=False)
```

```python
run_names_to_use
```

```python
mtr_df
```

```python
mtr_df.filter(like='roc')
```

```python
mtr_df.sort_values(by='delaney_test_rmse').iloc[:15]
```

```python
regr_tasks = ['delaney', 'bace_regression', 'clearance', 'lipo']
clf_tasks = ['bbbp', 'bace_classification', 'clintox', 'tox21']
```

```python
# we only want a single representative model per task
best_run_names = []
tasks = ['5M-MTR', '10M-MTR', '77M-MTR']
for task in tasks:
    best_run_names.append(mtr_df[mtr_df.pretraining_task == task].sort_values(by='min_eval_loss').iloc[0].run_name)
```

```python
df = mtr_df[mtr_df['run_name'] == "run_45"]
```

```python
df
```

```python
sns.set_context('talk')
sns.set_style('white')
sns.set(rc={'figure.figsize': (10, 8)})

```

```python
for regr_task in regr_tasks:
    fig, ax = plt.subplots(1, 2, figsize=(15, 6))
    for i,metric_name in enumerate(["rmse", "pearsonr"]):
        sns.barplot(
            data=df,
            x='pretraining_task',
            y=f'{regr_task}_test_{metric_name}',
            ci='sd', capsize=0.2, errwidth=3, palette='Blues',
            order=tasks,
            ax=ax[i]
        )
        ax[i].set_xlabel('Pretraining Task')
        ax[i].set_ylabel(f"{' '.join(metric_name.split('_')).title()}")
        ax[i].set_title(f"{' '.join(regr_task.split('_')).title()}")
#     plt.savefig(f"./barplots/{regr_task}_barplot.png", dpi=200)
```

```python
mm = {
    "roc_auc_score": "ROC AUC",
    "average_precision_score": "PR AUC",
}
for clf_task in clf_tasks:
    fig, ax = plt.subplots(1, 2, figsize=(15, 6))
    for i,metric_name in enumerate(mm.keys()):
        sns.barplot(
            data=df,
            x='pretraining_task',
            y=f'{clf_task}_test_{metric_name}',
            ci='sd', capsize=0.2, errwidth=3, palette='Reds',
            order=tasks,
            ax=ax[i]
        )
        ax[i].set_xlabel('Pretraining Task')
        ax[i].set_ylabel(f"{mm[metric_name]}")
        ax[i].set_title(f"{' '.join(clf_task.split('_')).title()}")
#     plt.savefig(f"./barplots/{clf_task}_barplot.png", dpi=200)
```

```python
fig, axes = plt.subplots(len(regr_tasks), 2, figsize=(15, 6*len(regr_tasks)))
for r,regr_task in enumerate(regr_tasks):
    for i,metric_name in enumerate(["rmse", "pearsonr"]):
        sns.barplot(
            data=mtr_df,
            x='run_name',
            y=f'{regr_task}_test_{metric_name}',
            ci='sd', capsize=0.1, errwidth=2, palette='Blues',
#             order=tasks,
            ax=axes[r, i],
#             hue='run_name',
            hue='pretraining_task'
        )
        axes[r, i].legend(loc=2, bbox_to_anchor=(1,1))
#         axes[r, i].set_xlabel('Pretraining Task')
        axes[r, i].set_xlabel('Model Configuration')
        axes[r, i].set_ylabel(f"{' '.join(metric_name.split('_')).title()}")
        axes[r, i].set_title(f"{' '.join(regr_task.split('_')).title()}")
        plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=1.0, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
# plt.savefig(f"./barplots/molnet_regr_all_runs_barplot.png", dpi=200)
plt.savefig(f"./barplots/molnet_regr_all_runs_barplot_2.png", dpi=200, bbox_inches='tight')
```
